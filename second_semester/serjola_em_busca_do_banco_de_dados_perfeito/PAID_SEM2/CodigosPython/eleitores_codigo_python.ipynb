```python
import pandas as pd
import numpy as np
```

```python
e_df = pd.read_csv('dados_eleitores_df.csv', sep=';', encoding = 'latin1')
```

```python
e_df.DS_ESTADO_CIVIL.unique()
```

```python
e_df['contador'] = 1 
display(e_df[['DS_ESTADO_CIVIL', 'contador']].loc[ e_df.DS_GENERO == 'MASCULINO'].groupby('DS_ESTADO_CIVIL').sum().reset_index())
display(e_df[['DS_ESTADO_CIVIL', 'contador']].loc[ e_df.DS_GENERO == 'FEMININO'].groupby('DS_ESTADO_CIVIL').sum().reset_index())
```

```python
display(e_df[['DS_ESTADO_CIVIL', 'contador', 'NR_ZONA']].loc[ e_df.DS_GENERO == 'MASCULINO'].groupby(['NR_ZONA', 'DS_ESTADO_CIVIL']).sum().reset_index())
display(e_df[['DS_ESTADO_CIVIL','contador', 'NR_ZONA']].loc[ e_df.DS_GENERO == 'FEMININO'].groupby(['NR_ZONA', 'DS_ESTADO_CIVIL']).sum().reset_index())
```
